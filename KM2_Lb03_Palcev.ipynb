```python
 import numpy as np
 import matplotlib.pyplot as plt
```

```python
def y(x):
    return np.sin(x)**4
```

```python
x_min = 1.; x_max = 5.; step = 0.05 
x_array = np.arange(x_min, x_max, step)
y_array = y(x_array)
matrix = np.transpose([x_array, y_array])
```

```python
fig1 = plt.figure()
```

```python
y_min, y_max = np.min(y_array), np.max(y_array)
ax1 = plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])
```

```python
line1, = ax1.plot([],[],'b') 
print(type(line1))
line1.get_xdata(), line1.get_ydata()
```

```python
def at_frame1(point): 
 x_coord = list(line1.get_xdata())
 y_coord = list(line1.get_ydata())
 x_coord.append(point[0])
 y_coord.append(point[1])
 line1.set_data(x_coord, y_coord) 
```

```python
fig1 = plt.figure()
ax1 = plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])
line1, = ax1.plot([], [],'g')
FuncAnimation(fig1, at_frame1, frames=matrix, repeat=False, interval=15)
```

```python
plt.close('all')
```

```python
def x1(t):
    return np.log(1 / np.sqrt(1 - t**4))

def y1(t):
    return  np.arcsin((1 - t**2) / (1 + t**2))
```

```python
t_array = np.arange(0, 0.99, 0.01)
x_array1 = x1(t_array)
y_array1 = y1(t_array)
print(y_array1)
matrix1 = np.transpose([x_array1, y_array1])
```

```python
x1_min, x1_max, y1_min, y1_max = np.min(x_array1), np.max(x_array1), np.min(y_array1), np.max(y_array1)
fig2 = plt.figure() 
ax2 = plt.axes()
plt.axis([x1_min-0.5, x1_max+0.5, y1_min-0.5, y1_max+0.5])
```

```python
def at_frame2(point):
    x_coord = list(line2.get_xdata())
    y_coord = list(line2.get_ydata())
    x_coord.append(point[0])
    y_coord.append(point[1])
    line2.set_data(x_coord, y_coord) 
```

```python
fig2 = plt.figure() 
ax2 = plt.axes()
plt.axis([x1_min-0.5, x1_max+0.5, y1_min-0.5, y1_max+0.5])

line2, = ax2.plot([],[],'b')

FuncAnimation(fig2, at_frame2, frames=matrix1, repeat=False, interval=15)
```

```python
plt.close('all')
```

```python
def at_frame2(point):
    """задает объект line2 точкой с координатами (point[0],point[1])
    Arguments :
    point : массив из двух элементов
    Returns : None
    """
    line2.set_data([point[0]],[point[1]])
```

```python
fig2 = plt.figure()
ax2 = plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])

def init():
    global line2
    line1, = ax2.plot(x_array,y_array, 'b')
    line2, = ax2.plot([],[],'go')
    plt.legend([r'$y(x) = \sin(x)^4$','Мovable point'])
    
FuncAnimation(fig2, at_frame2, frames=matrix, init_func=init, repeat=False, interval=20)
```

```python
plt.close('all')
```

```python
fig3 = plt.figure()
ax3 = plt.axes()
plt.axis([x_min-1, x_max+1, y_min-1, y_max+1])
def init():
    global point_B, secant_line # переменные сделаны глобальными, чтобы они были

    curve, = ax3.plot(x_array, y_array,'b') # графический объект для исходной фу

    A = np.array([x_array[0], y_array[0]])
    point_A, = ax3.plot(A[0],A[1],'ro') # графический объект для неподвижной точ

    B = np.array([x_array[-1], y_array[-1]])
    point_B, = ax3.plot(B[0],B[1],'go') # графический объект для подвижной точки

    secant_p = [A + (B - A)*t for t in [-2, 2]] # две точки секущей прямой, прох
    secant_p = np.array(secant_p)
    secant_line, = ax3.plot(secant_p[:,0], secant_p[:,1],'g') # графический объе

    plt.legend([r'$y(x) = \sin(x)^4$','Unmovable point A','Мovable point B','Sec'])
def at_frame3(point):
    global point_B, secant_line
    A = np.array([x_array[0], y_array[0]])
    B = np.array([point[0],point[1]])
    point_B.set_data([point[0]],[point[1]])
    
    if B[0]==A[0]:
        t = sympy.Symbol('t')
        y_x=sympy.sin(t)**4
        y_difx = sympy.diff(y_x, t) 
        k=y_difx.subs(t, A[0])
        y_kasstart = np.array([A[0],A[1]])
        y_kasend = np.array([A[0]+1,k+A[1]])
        y_kas=[y_kasstart+ (y_kasend-y_kasstart)*t for t in [-8, 8]]
        y_kas = np.array(y_kas)
        secant_line.set_data(y_kas[:,0], y_kas[:,1])
    else:
        secant_p = [A + (B - A)*l for l in [-1, 2]] # две точки секущей прямой, прох
        secant_p = np.array(secant_p)
        secant_line.set_data(secant_p[:,0], secant_p[:,1])
        
FuncAnimation(fig3, at_frame3, frames=matrix[::-1], init_func=init, repeat=False, interval=20)
```

```python

fig33 = plt.figure()
ax33 = plt.axes()
plt.axis([x1_min-0.5, x1_max+0.5, y1_min-0.5, y1_max+0.5])

def init1():
    """Создает начальное состояние графической области"""
    global line22 

    line11, = ax22.plot(x_array1, y_array1,'b')
    line22, = ax22.plot([], [], 'go')
    plt.legend(['-', 'Movable point'])
FuncAnimation(fig22, at_frame22, frames=matrix1, init_func=init1,repeat=False, interval=20)

def init():
    global point_B1, secant_line1 # переменные сделаны глобальными, чтобы они были

    curve1, = ax33.plot(x_array1, y_array1,'b') # графический объект для исходной фу

    A1 = np.array([x_array1[0], y_array1[0]])
    point_A1, = ax33.plot(A1[0],A1[1],'ro') # графический объект для неподвижной точ

    B1 = np.array([x_array1[-1], y_array1[-1]])
    point_B1, = ax33.plot(B1[0],B1[1],'go') # графический объект для подвижной точки

    secant_p1 = [A1 + (B1 - A1)*t for t in [-2, 2]] # две точки секущей прямой, прох
    secant_p1 = np.array(secant_p1)
    secant_line1, = ax33.plot(secant_p1[:,0], secant_p1[:,1],'g') # графический объе

    plt.legend([r'$y(x) = \sin(x)^4$','Unmovable point A','Мovable point B','Sec'])
def at_frame33(point):
    global point_B1, secant_line1
    A1 = np.array([x_array1[0], y_array1[0]])
    B1 = np.array([point[0],point[1]])
    point_B1.set_data([point[0]],[point[1]])
    
    if B1[0]==A1[0]:
        t = sympy.Symbol('t')
        y_x=sympy.asin((1 - t**2) / (1 + t**2))
        y_difx = sympy.diff(y_x, t) 
        k=y_difx.subs(t, A1[0]+0.01)
        y_kasstart = np.array([A1[0],A1[1]])
        y_kasend = np.array([A1[0]+0.01,k+A1[1]])
        y_kas=[y_kasstart+ (y_kasend-y_kasstart)*t for t in [-8, 8]]
        y_kas = np.array(y_kas)
        secant_line1.set_data(y_kas[:,0], y_kas[:,1])
    else:
        secant_p1 = [A1 + (B1 - A1)*l for l in [-2, 2]] # две точки секущей прямой, прох
        secant_p1 = np.array(secant_p1)
        secant_line1.set_data(secant_p1[:,0], secant_p1[:,1])
        
FuncAnimation(fig33, at_frame33, frames=matrix1[::-1], init_func=init, repeat=False, interval=20)
```
