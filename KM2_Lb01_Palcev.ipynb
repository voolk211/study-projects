## Задание 1.1. Задача о падении тела  

 Описание процесса. Тело запущено под углом $\alpha$ к горизонту на высоте $h_{start}$ и через время $t=T$ достигает другое тело, расположенное на расстоянии $s_{end}$ и высоте $h_{end}$

**Предположения**.
- Тело будем считать материальной точкой массы $m$.
- Движение запущенного тела осуществляется только под действием силы тяжести ${\bf F} = m(0,-g),$ где $m$ -- масса тела, $g$ -- ускорение свободного падения.
- Сопротивлением среды пренебрегаем.

**Данные**
- Задача рассматривается при следующих значениях параметров: $h_{start} = 1$ m, $h_{end} = 3$ m, $T = 3$ s, $s_{end} = 5$ m.
-  Угол запуска тела $\alpha$ не известен.
- Начальный момент времени движения тела полагаем равным нулю:$t = 0$
-  Значение массы тела $m$ не повлияет на функции перемещения тела.
  
**Задания для выполнения**
- $\color{red}\text{Определите}$ функцию перемещения тела $(s_{x}(t), s_{y}(t))$ под действием силы тяжести, где $s_{x}(t)$ и $s_{y}(t)$ — горизонтальная и вертикальная составляющая положения тела в момент времени $t$, соответственно.
- $\color{red}\text{Найдите}$ угол запуска тела $\alpha$.
- $\color{red}\text{Постройте}$ график траектории движения тела $(s_{x}(t), s_{y}(t))$ для $t\in[0,T]$.

Функция arange генерирует массив значений с указанным шагом, включая стартовое значение и исключая конечное. Имеет синтаксис: np.arange(start, end, step) и возвращает объект типа ndarray.

Аналитический вид функций получен следующим образом:
Для оси Ox: Объект двигается вдоль оси x без изменения скорости, поскольку на него действует только сила тяжести, действующая по оси y. В этом случае ускорение будет равно нулю, а значит и движение будет равномерным.
$x(t)=x_{0}+v_{0}*t+(0*t^2)/2 = x_{0}+v_{0}*t$

Для оси Oy: Объект двигается вдоль оси y с изменением скорости, поскольку на него действует Сила тяжести, действующая как раз по оси y. В этом случае ускорение будет равно ускорению свободного падения (-g), а значит и движение будет равноускоренным.
$y(t)=y_{0}+v_{0}*t+(a*t^2)/2 = y_{0}+v_{0}*t+(-g*t^2)/2$

Если начальный момент времени $t=t^*$ (не равен нулю), то уравнения для перемещения по осям x и y изменятся таким образом, что в них будет учитываться смещение по времени:

$x(t)=x_{0}+v_{0}*(t-t^*)$

$y(ty_{0}+v_{0}*(t-t^*)+(-g*(t-t^*)^2)/2$

## Задание 1.2. Определение момента времени касания телом земли 

```python
t = sympy.symbols('t',positive = True)
result = sympy.solve(h_start+v0_y*t-g*t**2/2,t)
t_star_exact = result[0]
t_star_exact
```

symbols создает и возвращает символьные объекты(кортеж таких объектов) для символьных вычислений в sympy. Выхов функции выглядит так: sympy.symbols(names, **kwargs), где kwargs - необязательный параметр задающий свойства символов: real = True, positive = true и т.д.

solve решает уравнения или системы. Выхов функции выглядит так: sympy.solve(equations, variables=None, **kwargs), где kwargs - необязательный параметр задающий дополнительные параметры:
dict = True - возвращает решения в виде словарей
set = True - возвращает решения в виде множества
check = True/False - проверяет корни или нет

```python
dh1 = sympy.symbols('dh1')
res1 = sympy.solve(dh1**2-4*dh1+5)
print(res1)

dh2 = sympy.symbols('dh2')
res2 = sympy.solve([dh1*5 - dh2*3 + 23, dh1*4 + dh2*3 -4], (dh1, dh2))
res2
```

```python
t_list= [i*step for i in range(int(1.5*T/step))]
s_y = [h_start + v0_y*t - g*t**2/2 for t in t_list]
for k,y in enumerate(s_y):
    if (y)<0:
        x1=(k-1)*step
        x2=k*step
        y1=s_y[k-1]
        y2=y
        break;

k = (y2 - y1) / (x2 - x1)
b = y1 - k * x1
t_zero = -b / k
print("Приближенное значение: ",end=" ")
print(t_zero)
print("Разница между точным и приближенным значением: ", end=" ")
print(t_star_exact - t_zero)
```
